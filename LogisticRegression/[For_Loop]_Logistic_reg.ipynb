```python
w1,w2,w3,b = 0,0,0,0

dw1,dw2,dw3,db = 0,0,0,0

da, dz = 0,0
```

```python
import numpy as np

def sigmoid(z):
  return 1/(1+np.exp(-z))

def update_param(var,dvar,lr=0.005):
  return var - lr * dvar

def binary_cross_entropy(a,y):
  return -((y * np.log(a)) + ((1-y) * np.log(1-a)))
```

```python
binary_cross_entropy(0.99,1)
```

```python
x1 = X.tenure
x2 = X.TotalCharges
x3 = X.PaperlessBilling

y = y.Churn

# x2 was string type before
x2 = pd.to_numeric(x2)

loss = 0
```

```python
m = len(x1)
```



---



```python
# after 10 epochs

epoch = 1000
```

```python
import csv

csv_file = open('report.csv','w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['epoch','w1','w2','w3','b','dw1','dw2','dw3','db','loss'])
```

```python
from time import time
```

```python
tic = time()

current_epoch = 0

while current_epoch <= epoch :

  loss = 0

  #Per epoch
  for i in range(m):
    
    z = w1*x1.iloc[i] + w2*x2.iloc[i] + w3*x3.iloc[i] + b
    a = sigmoid(z)

    loss += binary_cross_entropy(a,y.iloc[i])

    dz = a - y.iloc[i] 

    # Cumulative d
    dw1 += dz * x1.iloc[i]
    dw2 += dz * x2.iloc[i]
    dw3 += dz * x3.iloc[i]
    db += dz


  # (Cost function) : Average the loss
  dw1 /= m
  dw2 /= m
  dw3 /= m
  db  /= m

  # Update parameter (learning rate = 0.05)
  w1 = update_param(w1,dw1,lr=10**-6)
  w2 = update_param(w2,dw2,lr=10**-6)
  w3 = update_param(w3,dw3,lr=10**-6)
  b = update_param(b,db,lr=10**-6)

  csv_writer.writerow([current_epoch,w1,w2,w3,b,dw1,dw2,dw3,db,loss])
  #csv_writer.writerow([current_epoch,w1,w2,w3,b,dw1,dw2,dw3,db,loss])

  print(f'Epoch : {current_epoch}/{epoch}')
  current_epoch += 1



toc = time()

print(f"{epoch} epoch : {toc-tic} second")
```

```python
print(f"{epoch} epoch : {(toc-tic)/60} minute")
```

```python
csv_file.close()
```
